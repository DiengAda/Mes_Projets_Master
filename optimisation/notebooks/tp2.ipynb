**Question 2:**  

Methode du gradiant projeté

Mise en oeuvre de la méthode du gradient projeté pour résoudre le problème de minimisation sous contraintes:
$$
\inf_{(x,y)\in Q} J(x,y)
$$


La méthode du gradient projeté est une méthode d'optimisation qui permet de résoudre des problèmes de minimisation sous contraintes en projetant itérativement les itérés sur la contrainte. Pour résoudre le problème de minimisation $\inf_{(x,y)\in Q} J(x,y)$, nous pouvons appliquer la méthode du gradient projeté en suivant les étapes suivantes :

1. Choisir un point initial $\bar{X}_0$ qui satisfait la contrainte $Q$ (par exemple, $\bar{X}_0=(0,0)$).
2. Calculer le gradient de la fonction objectif $J(x,y)$ en $\bar{X}_k$.
3. Projeter $\bar{X}_k - \alpha_k \nabla J(\bar{X}_k)$ sur la contrainte $Q$ pour obtenir le nouvel itéré $\bar{X}_{k+1}$ :
$$
\bar{X}_{k+1} = \operatorname{Proj}_Q(\bar{X}_k - \alpha_k \nabla J(\bar{X}_k)),
$$
où $\operatorname{Proj}_Q$ est l'opérateur de projection sur la contrainte $Q$.
4. Si la norme de la différence entre $\bar{X}_{k+1}$ et $\bar{X}_k$ est suffisamment petite (en dessous d'un seuil de tolérance), arrêter l'algorithme et renvoyer $\bar{X}_{k+1}$ comme solution approchée. Sinon, retourner à l'étape 2 en utilisant $\bar{X}_{k+1}$ comme nouvel itéré.


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Fonction objective
def J(x):
    res = 2*x[0]**2 + 3*x[0]*x[1] + 2*x[1]**2
    return res

# Gradient de la fonction objective
def gradJ(x):
    gr = np.array([4*x[0] + 3*x[1], 3*x[0] + 4*x[1]])
    return gr

#Projection sur l'ensemble Q={x | x[0] < -1/2, x[1] < -1/2}
def projQ(x):
    pr = np.array([min(x[0], -1/2), min(x[1], -1/2)])
    return pr
```

```python
# Algorithme de gradient projeté
def gradient_projete(x0, rho, tol):
    x = x0.copy()
    iteres = [x]
    while True:
        grad = gradJ(x)
        d = -grad
        x_next = projQ(x + rho * d)
        iteres.append(x_next)
        if np.linalg.norm(x_next - x) < tol:
            break
        x = x_next
    return x, iteres
```

Représenter les itérés obtenus à l’aide de cette méhode

```python
def plot_iteres(iteres):
    x = np.linspace(-1, 1, 100)
    y = np.linspace(-1, 1, 100)
    X, Y = np.meshgrid(x, y)
    Z = J([X, Y])
    plt.contour(X, Y, Z, 50)
    plt.plot([x[0] for x in iteres], [x[1] for x in iteres], '*')
    plt.show()
    

 
```

```python
x0 = np.array([-0.6, 0.2])
rho = 1e-1
tol = 1e-3
x, iteres = gradient_projete(x0, rho, tol)
print("Solution: ", x)
print("Valeur de la fonction objective: ", J(x))
print("Nombre d'itérations: ", len(iteres))
plot_iteres(iteres)
```

```python

# Définition des fonctions de pénalité phi1(x, eps) et phi2(x, eps)
def phi1(x, eps1):
    return (1/eps1) * (-1/2 - x[0])**2

def phi2(x, eps2):
    return (1/eps2) * (-1/2 - x[1])**2

# Définition de la fonction de pénalisation
def penalized_func(x, eps3):
    P1 = phi1(x, eps3)
    P2 = phi2(x, eps3)
    return J(x) + eps3 * (P1 + P2)
```

```python
# Définition des gradients des fonctions de pénalité phi1(x, eps) et phi2(x, eps)
def grad_phi1(x, eps4):
    return np.array([-2*(1/eps4)*(-1/2 - x[0]), 0])

def grad_phi2(x, eps5):
    return np.array([0, -2*(1/eps5)*(-1/2 - x[1])])

# Définition du gradient de la fonction de pénalisation
def grad_penalized_func(x, eps6):
    GradJ = gradJ(x)
    gradP1 = grad_phi1(x, eps6)
    gradP2 = grad_phi2(x, eps6)
    return GradJ + eps6 * (gradP1 + gradP2)
```

```python
## methode du gradient a pas fixe
def gradient_fixe(x0, eps, step_size, tole):
    x = x0.copy()
    iteres = [x]
    while True:
        grad = grad_penalized_func(x, eps)
        x_next = x - step_size * grad
        iteres.append(x_next)
        if np.linalg.norm(x_next - x) < tole:
            break
        x = x_next
    return x, iteres

```

2. "np.transpose(R) @ R" : Calcule le produit de la matrice R avec sa transposée. Cela donne une matrice symétrique de taille n x n.

3. "0.05 * (np.transpose(R) @ R)" : Multiplie la matrice symétrique obtenue à l'étape précédente par un facteur scalaire de 0.05. Ce facteur détermine l'ampleur de la perturbation qui sera ajoutée à la matrice diagonale.

4. "A = A + 0.05 * (np.transpose(R) @ R)" : Ajoute la perturbation calculée à la matrice diagonale A. Ainsi, la matrice A finale représente la somme de la matrice diagonale initiale et de la perturbation.

Par consequnt on peut dire que la dernière ligne du programme introduit une corrélation entre les différentes actions en modifiant la matrice A. Cette corrélation est représentée par une matrice aléatoire R (et sa transposée), qui simule l'incertitude et la variabilité présentes sur le véritable marché boursier. Le facteur 0.05 est utilisé pour contrôler l'intensité de cette corrélation supplémentaire par rapport aux rendements attendus des actions, qui sont représentés par la matrice diagonale initiale A. Ainsi, cette ligne de code permet de prendre en compte à la fois les rendements individuels des actions et leur corrélation potentielle.

```python
def Uzawa(A, B, c, x0, rho):
    x = x0.copy()
    lam = np.zeros(B.shape[0])
    tol = 1e-6
    
    while True:
        gradien = A @ x
        lam = lam + rho * (B @ x - c)
        x = x - rho * (gradien + B.T @ lam)
        if np.linalg.norm(gradien + B.T @ lam) < tol:
            break
    return x, lam # lam est multiplicateur de Lagrange
         
```
